# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    if numero == 1 and palo.lower() == "espada":
        return 14
    elif numero == 1 and palo.lower() == "basto":
        return 13
    elif numero == 7 and palo.lower() == "espada":
        return 12
    elif numero == 7 and palo.lower() == "oro":
        return 11
    elif numero == 3:
        return 10
    elif numero == 2:
        return 9
    elif numero == 1:  
        return 8
    elif numero == 12:
        return 7
    elif numero == 11:
        return 6
    elif numero == 10:
        return 5
    elif numero == 7:  
        return 4
    elif numero == 6:
        return 3
    elif numero == 5:
        return 2
    elif numero == 4:
        return 1
    else:
        print("Carta no valida")

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"Carta:{self.numero}, Palo:{self.palo}, Valor: {self.valor}"

carta = Carta(12, "copa")
print(carta)

```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random

class Mazo:
    
    def __init__(self):
        self.palos = ["oro", "espada", "basto", "copa"]
        self.numeros_cartas = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]
        self.mazo = [(numero, palo) for palo in self.palos for numero in self.numeros_cartas]

    def mezclar(self):
        random.shuffle(self.mazo)

    def repartir(self, cantidad):

        jugador = []
   
        for _ in range(cantidad):
            jugador.append(self.mazo.pop())

        return jugador
    
m = Mazo()
m.mezclar()
print(m.repartir(3))
print(m.repartir(3))

```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
class Jugador:
    def __init__(self, nombre):
        self.nombre=nombre
        self.cartas=[]
        self.puntos=0
    def recibir_cartas(self, cartas):
        for carta in cartas:
            numero,palo = carta
            self.cartas.append(Carta(numero,palo))
            
    
    def jugar_carta(self):
        random.shuffle(self.cartas)
        return self.cartas.pop()

j = Jugador("Juan")
j.recibir_cartas(m.repartir(3))
print(j.jugar_carta())
```
